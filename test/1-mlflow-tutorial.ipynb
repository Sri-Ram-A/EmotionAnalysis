```python
import mlflow
import mlflow.tensorflow
import warnings
warnings.filterwarnings("ignore")
from pathlib import Path
NOTEBOOK_DIR  = Path().resolve()
BASE_DIR = NOTEBOOK_DIR.parents[0]
DATASET_DIR = BASE_DIR /  "data" / 'TwitterSentimentAnalysisDataset' / "preprocessed"
DATASET_FILE_PATH = DATASET_DIR  / 'twitter_training.csv'
DATASET_FILE_PATH.exists()
print("Project base dir is :",BASE_DIR)
mlflow.enable_system_metrics_logging()
```

```python
mlflow.set_tracking_uri(f"http://localhost:5000")
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv(DATASET_FILE_PATH,usecols=["sentiment","cleaned_text"],nrows=100)
df["cleaned_text"] = df["cleaned_text"].astype(str)
print("Null values",df.isnull().sum())
print("Shape of df :",df.shape)
```

```python
mlflow.set_experiment("0-EXPLORE")
```

```python
# Print connection information
print(f"MLflow Tracking URI: {mlflow.get_tracking_uri()}")
print(f"Active Experiment: {mlflow.get_experiment_by_name('0-EXPLORE')}")
```

```python
run =  mlflow.start_run(run_name="Learning")
```

```python
# in UI- logged under Evaluation
mlflow.log_table(df.sample(50), artifact_file="artifact/table.json")
```

```python
import psutil, time, mlflow
# in UI - Logged under Chart and Table/x/Metrics/- 
for i in range(30):
    mlflow.log_metric("cpu_percent", psutil.cpu_percent())
    mlflow.log_metric("ram_percent", psutil.virtual_memory().percent)
    time.sleep(0.5)
```

```python
#  In UI - Logged under Table/x/Tags
mlflow.set_tag("description", "Baseline model with CountVectorizer + LR")
mlflow.set_tag("dataset", "cleaned_text_v1")
mlflow.set_tag("experiment_stage", "baseline")
```

```python
#  In UI - Logged under Table/x/Parameter
mlflow.log_params({
    "vectorizer": "CountVectorizer",
    "vocab_size": 100,
    "model": "LogisticRegression",
    "C": 1.0,
    "max_iter": 1000
})
```

```python
mlflow.end_run()
```
